```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load
file_to_load = "purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load) 
purchase_data
```

## Player Count

* Display the total number of players


```python
#count_player = print(f"The total number of players : {len(purchase_data['SN'].value_counts())}")
player_counts = len(purchase_data['SN'].value_counts())
total_player = pd.DataFrame({'Total Player':[player_counts]})
total_player
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_item = len(purchase_data['Item ID'].unique())
ave_price = purchase_data['Price'].mean()
num_of_purchase = purchase_data['Purchase ID'].count()
total_revenue = purchase_data['Price'].sum()

sum_table = pd.DataFrame({'Number of Unique Items':[unique_item],
                          'Average Price':[ave_price],
                          'Number of Purchases':[num_of_purchase],
                          'Total Revenue':[total_revenue]})
sum_table
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Drop Duplicates
df = purchase_data[['Gender', 'Purchase ID', 'SN']]
df = df.drop_duplicates('SN')

gender_counts = df.Gender.value_counts()
gender_percentage = gender_counts/player_counts

Gender_demo = pd.DataFrame({'Total Count':gender_counts,
                           'Percentage of Players':gender_percentage})
Gender_demo
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Groupby Gender
group_gender = purchase_data.groupby('Gender')

# Purchase count 
purchase_count = group_gender['Price'].count()

# Total purchase 
total_purchase = purchase_data.groupby(["Gender"])["Price"].sum()

# Average purchase price 
ave_price = purchase_data.groupby(["Gender"])["Price"].mean()

# Average purchase price per person 
ave_purchase_per_person = total_purchase/gender_counts

# Create dataframe 
purchase_analysis = pd.DataFrame({'Purchase Count':purchase_count,
                                 'Average Purchase Price':ave_price,
                                 'Total Purchase Value':total_purchase,
                                 'Avg Total Purchase per Person':ave_purchase_per_person})
purchase_analysis
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Select players and age from dataset 
age_demo = purchase_data[['SN','Age']]

# Drop duplicated values
age_demo = age_demo.drop_duplicates('SN')

# Creat bins and labels
bins = [0,9,14,19,24,29,34,39,99]
labels = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']

# Binning
age_demo['Total Count'] = pd.cut(age_demo['Age'], bins, labels = labels )
```

```python
# Calculate percentage of players

# Total count by each age group 
age_count = age_demo['Total Count'].value_counts()

# Percentage by age 
total_players = age_demo['SN'].count()
age_percentage = age_count/total_players 

summary_age_demo = pd.DataFrame({'Total Count':age_count,'Percentage of Players':age_percentage})

# Sort Index
summary_age_demo.sort_index()

# Formatting 
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Bining 
bins = [0,9,14,19,24,29,34,39,99]
labels = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']
purchase_data['Age Ranges'] = pd.cut(purchase_data['Age'], bins, labels = labels )

# Group by Age Range
purchase_analysis = purchase_data.groupby('Age Ranges')

# Calculate values
purchase_count = purchase_analysis.Price.count()
ave_purchase_price = purchase_analysis.Price.mean()
total_purchase_value = purchase_analysis.Price.sum()
Ave_total_per_person = total_purchase_value/summary_age_demo['Total Count']

# Summarize table
purchase_analysis_summary = pd.DataFrame({'Purchase Count':purchase_count,
                                         'Average Purchase Price':ave_purchase_price,
                                         'Total Purchase Value':total_purchase_value,
                                         'Average Total Purchase per Person':Ave_total_per_person})
purchase_analysis_summary

# Formatting 
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Retrieve Item ID, Item Name and Price data
item_analysis = purchase_data[['Item ID','Item Name','Price']]

# Group by Item ID and Item Name
item_analysis = item_analysis.groupby(['Item ID','Item Name'])

# Calulating Purchase Count, Item Price and Total Purchase Value
purchase_count = item_analysis['Price'].count()
total_purchase_value = item_analysis['Price'].sum()
item_price = total_purchase_value/purchase_count


item_analysis_summary = pd.DataFrame({'Purchase Count':purchase_count,
                                    'Item Price':item_price,
                                    'Total Purchase Value':total_purchase_value})

# Sort values by Purchase Count 
item_analysis_summary.sort_values("Purchase Count", ascending=False).head()
```
